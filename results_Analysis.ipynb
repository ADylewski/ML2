```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from IPython.display import Image, display
```

How many images in the validation set?

```python
directory = "datasets/cocoa_diseases/images/val"

# Count files only (exclude subdirectories)
num_files = sum(1 for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f)))

print(f"Number of files in '{directory}': {num_files}")
```

How many labelled objects in the validation set?

```python
directory = "datasets/cocoa_diseases/labels/val"
total_lines = 0

# Iterate over all .txt files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        with open(file_path, "r") as file:
            # Count the lines in the current file
            total_lines += sum(1 for _ in file)

print(f"Total number of lines in all .txt files: {total_lines}")
```

The number of objects in the labelled set is lower than the number of objects from the confusion matrix (289 < 419). 

```python
display(Image(filename='runs/detect/train/confusion_matrix.png', width=500))
```

However, this is normal, since the algorithm predicts boxes where no actual objects exist.

```python
# Load the images using matplotlib.image
im1 = mpimg.imread('runs/detect/train/val_batch0_labels.jpg')
im2 = mpimg.imread('runs/detect/train/val_batch0_pred.jpg')

# Plot the images side by side
plt.figure(figsize=(10, 5))  # Adjust figsize as needed
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.imshow(im1)
plt.axis('off')  # Hide axes
plt.title("Labels")  # Add title for clarity

plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.imshow(im2)
plt.axis('off')  # Hide axes
plt.title("Predictions")  # Add title for clarity

plt.tight_layout()
plt.show()
```
